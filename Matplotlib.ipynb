```python
from mpl_toolkits.mplot3d import Axes3D

# Create some sample data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Create a 3D figure
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the data
ax.plot_surface(X, Y, Z)

# Add labels and title
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
ax.set_title('3D Surface Plot')
plt.show() 
```

```python
# Create some sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a figure and axis objects
fig, ax = plt.subplots()

# Plot the data
ax.plot(x, y)

# Add labels and title
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_title('2D Line Plot')

# Show the plot
plt.show()
```

```python
# Create some sample data
labels = ['Apples', 'Oranges', 'Bananas', 'Grapes']
values = [30, 45, 12, 70]

# Create a figure and axis objects
fig, ax = plt.subplots()

# Plot the data
ax.bar(labels, values)

# Add labels and title
ax.set_xlabel('Fruit')
ax.set_ylabel('Quantity')
ax.set_title('Fruit Quantities')

# Show the plot
plt.show()
```

```python
# Create some sample data
x = np.array(['A', 'B', 'C', 'D'])
y = np.array([3, 6, 8, 4])

# Create a figure and axis objects
fig, ax = plt.subplots()

# Plot the data
ax.vlines(x=x, ymin=0, ymax=y, color='red', linewidth=3)
ax.scatter(x, y, color='orange', s=200)

# Add labels and title
ax.set_xlabel('Category')
ax.set_ylabel('Value')
ax.set_title('Lollipop Chart')

# Show the plot
plt.show()
```

```python
# Create some sample data
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# Create a figure and axis objects
fig, ax = plt.subplots()

# Plot the data
parts = ax.violinplot(data)

# Change the color of the violin plot to blue
for pc in parts['bodies']:
    pc.set_facecolor('blue')

# Add labels and title
ax.set_xlabel('Data')
ax.set_ylabel('Value')
ax.set_title('Violin Plot')

# Show the plot
plt.show()
```
